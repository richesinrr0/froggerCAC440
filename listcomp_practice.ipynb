```python
#3. Count the number of spaces in a string (use string above):
string = "Practice Problems to Drill List Comprehension in Your Head."
space_count = 0
for i in string:
    if i == ' ':
        space_count+=1
print('space count: ', space_count)
```

```python
#4. Remove all of the vowels in a string (use string above):
string = "Practice Problems to Drill List Comprehension in Your Head."
for i in string:
    if i in list('aeiou'):
        string = string.replace(i,'')
string
```

```python
#5. Find all of the words in a string that are less than 5 letters (use string above):
string = "Practice Problems to Drill List Comprehension in Your Head."
string = string.split(' ')
for i in string:
    if len(i) < 5:
        print(i, end= ' ')
```

```python
#6. Use a dictionary comprehension to count the length of each word in a sentence (use string above):
string = "Practice Problems to Drill List Comprehension in Your Head."
word_lens = {key: len(key) for key in string.split()}
word_lens
```

```python
#7. Use a nested list comprehension to find all of the numbers from 1–1000 that are divisible by 
#any single digit besides 1 (2–9):
nums = [i for i in range(1,1001) for j in range(2,10) if i//j==0]

```

```python
#8. For all the numbers 1–1000, use a nested list/dictionary comprehension to find the highest single digit any of the 
#numbers is divisible by
nums = [max([i for i in range(1,10) if j % i == 0]) for j in range(1,1001)]
nums
```

```python
def ispalindrome(word):
    if len(word) < 2: return True
    if word[0] != word[-1]: return False
    return ispalindrome(word[1:-1])
```
